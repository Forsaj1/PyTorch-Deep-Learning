### Matrix multiplication

Two main ways of performing in neural networks and deep learning
1. Element-wise multiplication
2. Matrix multiplication (dot product)

https://www.mathsisfun.com/algebra/matrix-multiplying.html

There are 2 main rules performing Matrix multiplication:
1. The **inner dimensions** must match:
* `(3, 2) @ (3, 2)` won't work
* `(2, 3) @ (3, 2)` will work
* `(3, 2) @ (2, 3)` will work

2. The resulting matrix has the shape of the **outer dimensions**:
* `(2, 3) @ (3, 2)` -> `(2, 2)`
* `(3, 2) @ (2, 3)` -> `(3, 3)`

http://matrixmultiplication.xyz/

````python
# Element-wise multiplication
print(tensor, '*', tensor)
print(f"Equals: {tensor*tensor}")
````

````python
# Matrix multiplication
%%time
torch.matmul(tensor, tensor)
````

````python
# Shapes for Matrix multiplications
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])
tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]])
````

````python
# torch.mm() is an alias for torch.matmul()
torch.mm(tensor_A, tensor_B.T)
````

````python
# To fix tensor shape issues use Transpose
tensor_B.T
````

````python
# The natrix multiplication operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimesnions must match")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
````

````python
output_2 = torch.matmul(tensor_A.T, tensor_B)
output_2
````

## Reshaping, stacking, squeezzing and unsqueezing tensors

* Reshaping - reshapes an input tesbor to a defined shape
* View - return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine multiple tesnors on top of each other (vstack) or side by side (hstack)
* Squezze - removes all `1` dimensions from a tensor
* Unaqueeze - adds a `1` dimension to a target tensor
* Permute - return a view of the input w/ dimensions permuted (swapped) is a certain way

````python
# Create a tensor
x = torch.arange(1., 10.)
x, x.shape
````

````python
#Add an extra dimension
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
````

````python
# Change the view
z = x.view(1, 9)
z, z.shape
````

````python
# Changing z changes x (because a view of a tensor shares the same memory with the original tensor)
z[:, 0] = 5
z, x
````

````python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
````

````python
# Squeeze - removes all single dimensions from a target tensor
print(f"Previous tensor : {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimesnion from x_reshaped
x_squeezed = torch.squeeze(x)
print(f"\nNew tensor : {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
````

````python
# Unsqueeze - adds a single dimension to a target tensor at a specific dimension (dim)
print(f"Previous target : {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

# Add an extra dimesnion from x_reshaped
x_unsqueezed = torch.unsqueeze(x_squeezed, dim=0)
print(f"\nNew tensor : {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
````

````python
# Permute - rearranges the dimensions of a taget tensor in a specified order
x_original = torch.rand(size=(224, 224, 3)) # [height, width, colour_channels]

# Permute the original tensor to rearrange the axis (or dim) order
x_permuted = torch.permute(x_original, (2, 0, 1)) # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}") # [colour_channels, height, width]
````

````python
x_original[0, 0, 0] = 9592
x_original[0, 0, 0], x_permuted[0, 0, 0]
````

## PyTorch tensors & NumPy

NumPy is a populat scientific Python numerical computing lib
and because of this, PyTorch has functionality to interact with it.

* Data in NumPy, want in PyTorch tensor -> `torch.from_numpy(ndarray)`
* PyTorch tensor to Numpy -> `torch.Tensor.numpy()`

````python
# Numpy array to tensor
import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array) # warning: when converting from numpy to pytorch, pytorch reflects numpy's default dtype as float64, unless specified otherwise
array, tensor
````

````python
# Change the value of array, what will this do to tensor?
array = array + 1
array, tensor
````

````python
# Tensor to numpy change
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
````

````python
# Change the tensor, what happens to numpy array?
tensor = tensor + 1
tensor, numpy_tensor
````

## Reproducibilty (trying to take random out of random)

How a neural network learns:

`start w/ random numbers -> tensor operations -> update random numbers to try and make them better representations of the data -> again -> again -> ...`

To reduce the randomness in neural networks, PyTorch comes with the concept of a **random seed**

Essentially what the random seed does is "flavour" the randomness

https://pytorch.org/docs/stable/notes/randomness.html?highlight=reproducibility

````python
# Create 2 random tensors
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
````

````python
# Let's make some random but reproducible tensors
RANDOM_SEED = 69

torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
````
